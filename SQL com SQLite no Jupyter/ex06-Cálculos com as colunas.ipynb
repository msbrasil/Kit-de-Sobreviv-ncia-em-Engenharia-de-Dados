É possível fazer operações básicas de matemáticas e até calcular estatísticas com as colunas que contém números.

```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect('../primeiro_banco')
cur = con.cursor()
print('Conectado ao banco de dados')
```

Consultar quais tabelas existem no banco de dados e colunas

```python
query = 'SELECT name FROM sqlite_master WHERE type="table"'

df = pd.read_sql_query(query, con)
print(df)
```

```python
query = 'SELECT * FROM Customers LIMIT 5'

df = pd.read_sql_query(query, con)
df
```

É possivel testar os operadores com o SELECT 

```python
query = 'SELECT 3 + 4' #Adição

df = pd.read_sql_query(query, con)
print(df)

print()
query = 'SELECT 12 / 4' #Subtração

df = pd.read_sql_query(query, con)
print(df)

print()
query = 'SELECT 12 % 4' #Módulo

df = pd.read_sql_query(query, con)
print(df)

print()
query = 'SELECT ROUND(17465.4444445298)' #Arredondamento

df = pd.read_sql_query(query, con)
print(df)

print()
query = 'SELECT POWER(2,3)' #Quociente da divisão

df = pd.read_sql_query(query, con)
print(df)
```

Operações com colunas 

```python
#tratamento da coluna AnnualIncome que estava como string.
con.executescript(
'''
UPDATE
    Customers
SET
    AnnualIncome = CAST(REPLACE(REPLACE(AnnualIncome, ",",""), "$","") AS INT)
'''
)

con.commit()
```

```python
query = 'SELECT * FROM Customers LIMIT 5'

df = pd.read_sql_query(query, con)
df
```

```python
query = '''SELECT FirstName, 
LastName, 
Occupation, 
AnnualIncome AS INT,
TotalChildren,
AnnualIncome / (TotalChildren + 2) AS AvgIncome
FROM Customers 
WHERE MaritalStatus = "M"
'''

df = pd.read_sql_query(query, con)
df
```

É possível usar calculos na condição `WHERE`

```python
query = '''SELECT FirstName, 
LastName, 
Occupation, 
AnnualIncome AS INT,
TotalChildren,
AnnualIncome / (TotalChildren + 2) AS AvgIncome
FROM Customers 
WHERE MaritalStatus = "M" AND AnnualIncome / (TotalChildren + 2) > 20000
'''

df = pd.read_sql_query(query, con)
df
```

Estatísticas com funções de agregação

```python
query = 'SELECT AVG(AnnualIncome) FROM Customers' #média

df = pd.read_sql_query(query, con)
df
```

```python
query = 'SELECT SUM(AnnualIncome) FROM Customers' #soma

df = pd.read_sql_query(query, con)
df
```

```python
query = 'SELECT MIN(AnnualIncome), MAX(AnnualIncome) FROM Customers' #mínimo e máximo

df = pd.read_sql_query(query, con)
df
```

```python
query = 'SELECT COUNT(FirstName) FROM Customers' #contagem

df = pd.read_sql_query(query, con)
df
```
